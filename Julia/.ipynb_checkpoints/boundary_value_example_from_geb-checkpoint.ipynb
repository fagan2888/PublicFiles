```julia
using Pkg
using Plots
using DifferentialEquations
using Distributions
using BoundaryValueDiffEq
using SpecialFunctions
```

```julia
function dummy_example_from_paper!(du, u, p, v_2)
             v_1, b = u;
             F1overf1(x) = x # Corresponding to F1=x
             F2overf2(x) = (sqrt(pi)*erf(x)./exp(-x.^2)/2);  # Corresponding to F2=c erf(x) (truncated normal)
        if v_2 == 1.0 #v_2_upper
            du[1] = 0
            du[2] = 0
        elseif v_2 == 0.0
            du[1] = 0
            du[2] = 0
        elseif v_2 == b
            du[1] = 0
            du[2] = 0
        else
            du[1] = F1overf1(v_1) / F2overf2(v_2) * (v_1 - b) / (v_2 - b)
            du[2] = (v_1 - b) / F2overf2(v_2)
        end
end
function boundary_conditions!(residual, u, p, t)
    residual[1] = u[end][1] - 1.0
end
v_init = [0.0,0.0]
v_span = (0.001,1.0)
bvp_twopoint = TwoPointBVProblem(dummy_example_from_paper!, boundary_conditions!, v_init, v_span)
sol = solve(bvp_twopoint, MIRK4(), dt=0.001) ## plot the solution as it comes
```

```julia
v_1 = [x[1] for x in sol.u]
v_2 = sol.t
b = [x[2] for x in sol.u]

## Plot the solution in terms of b(v) for each bidder
plot(b, v_1, label = "b(v_1)")
plot!(b, v_2, label = "b(v_2)")
```

# Auction Equilibriums

Suppose you have an auction for an item: a seller has an object of value and asks bidders to each submit a bid.
We're looking at "first price" auctions, which means that the bidder with the highest bid wins.The bids are obviously secret and submitted simultaneously.

We are looking at a single auction (one item at sale only). So bidding strategies, knowledge of resources and iterative game aspects can be ignored. We are only selling one item not multiple.

For each bidder $i$ the item provides a certain value $v_i$ where $v_i \in [a,b]$.
Each bidder submits the bid $b_i$. We assume the bid maximizes the expected utility $E[\pi]_i$.
$$E[\pi]_i = (v-b)*Pr_i[\forall i \neq j: b_i > b_j ]$$
$Pr_i[\forall i \neq j: b_i > b_j]$ or just $Pr[b_i]$ for short is the probability that the $b_i$ is the highest of all bids. This function is obviously uncomputable at the time one has to submit the sealed bid.

To continue the analysis and to simplify the math we assume these things:
 - There are two biders only
 - Each bider choose his bid $b_i$ to maximize his $E[\pi]_i$
 - The distribution which models each bidder $i$'s likelyhood to bid a certain value is a well behaved distribution over a bounded interval $[a,b]$ with cdf $G_i(x)$ 
 
This immediatly implies $Pr[b_1] = Pr[ b_1 > b_2 ] = G_2(b_1)$.  

The topic at hand however is Auction Equilibriums and not auctioning using time traveling informatoin. It has been proven that under certain necessary conditions $G_1 = G_2$ and we can therefor just recognize $G$. 
As a next steps we define $\phi(v), \xi(v)$ for bidder i as $v \mapsto \arg\max_b (v-b)*Pr_i[b_i > b_j ]$. A function that returns the "best bid" for each value $v$ that could be taken. We can use this definition to observe the following: $$ Pr(b > b') = Pr( b > \xi(v') ) = Pr(\phi^{-1}(b) > \phi^{-1}(\xi(v')) = Pr(v > \phi^{-1}(\xi(v'))) $$


If we have an equilibrium those assumptions must apply
 - The worst kind of bidder makes no money $(b-v)=0$
 - If some biders have differnt distributoins they would bid the smae thing
 
 ### Illustrating example:
 the value of A is distributed U[0,1] and the value of B is distributed U[0.3,0.8]. in order for any bidding strategy to make sense, it needs to be that phi(0) is the same as xi(0.3) and phi(1) is the same as xi(0.8). imagine that phi(0) < xi(0.3) and that is, if bidder A draws his 'worst type' he bids lower than if bidder B draws her 'worst type'.
